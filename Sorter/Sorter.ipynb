## Sorter 

Helper utility that makes it possible to sort two datasets with the same sentences but, different / messed up sent_ids that need to be aligned between the two datasets.

```python
# imports
import json
import os
```

```python
# makedirs if not exist
os.makedirs("../../etl/data/processed/OrderedOutput", exist_ok=True)
```

```python
# parameters
# two files with a list of dicts, each dict is a sentence and has a sent_id.
# the sent_ids differ in the two files (two sentences with the same sent_id are not the same sentence)
# this must be fixed

# the file by which must be sorted (in terms of sent_id)
SORT_BY="../../annex/own_datasets/tt_06_va_soft/results/test/perin/01_test_prediction.json"

# the file needs to be sorted
TO_SORT="../../etl/data/processed/Perin_Preprocessing/01_test.json" 
```

```python
def read_json(INP_FILE):
    with open(INP_FILE, encoding="utf-8") as f:
        data = json.load(f)
        return data

sort_by = read_json(SORT_BY)
to_sort = read_json(TO_SORT)
```

```python
import copy

to_sort_clean = []

for sent in sort_by:
    for sent2 in to_sort:
        if sent["text"] == sent2["text"]:
            sent_x = copy.deepcopy(sent2)
            sent_x["sent_id"] = sent["sent_id"]
            to_sort_clean.append(sent_x)
```

```python
def save_json(OUT_FILE, data):
    with open(OUT_FILE, "w", encoding="utf-8") as f:
        json.dump(data, f, indent=4, ensure_ascii=False)

save_json("../../etl/data/raw/OrderedOutput/01_test.json", to_sort_clean)
```
