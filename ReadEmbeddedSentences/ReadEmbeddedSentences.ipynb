```python
sents,rels={},{}
c=0
with open(RAW_DATA_PATH,"r") as f:
    for line in f:
        if line[0]=='#':
            continue
        # from IPython.core.debugger import Pdb; Pdb().set_trace()
        try:
            id,d=line.split("=",1) # id nicht verwenden, selber hochzählen mit c
        except ValueError as e:
            print(f"Failed at line {line}, with error {e}")
        sents[c]=d.strip("\n")
        _,rel=next(f).split(":")
        rels[c]=rel.strip("\n")
        c+=1
        #if c>9000: # nur fürs testen, dann weg
        #    break
```
